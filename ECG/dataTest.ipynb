```python
import pandas as pd
import numpy as np
import wfdb
```

```python
record_path = 'physionet.org/files/cebsdb/1.0.0/p002'
record = wfdb.rdrecord(record_path)
print(record.p_signal)        # NumPy 배열 (샘플 수 x 채널 수)
print(record.sig_name)        # 채널 이름
print(record.fs)              # 샘플링 주파수
```

```python
import matplotlib.pyplot as plt

# 첫 채널 플롯
plt.plot(record.p_signal[:10000, 0])
plt.title(record.sig_name[0])
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```
